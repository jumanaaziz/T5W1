## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
#scatter plot

sns.scatterplot(data=df, x='City', y='JamsDelay')
plt.title('Relation between City and JamDelay')
plt.xlabel('City')
plt.ylabel('Jam delay')
plt.show()
```

```python

#histogram
sns.displot(data= df, x='City', height=6, aspect=2)
plt.title('City and JamDelay')
plt.xlabel('City')
plt.ylabel('Jam delay')
plt.show()
```

```python
#heatmap

sns.heatmap(matrix, annot=True)
plt.show()
```

```python
#pie chart
count =df['JamsLength'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(count, labels=count.index, autopct='%1.1f%%', startangle=140)
plt.title('Jams count')
plt.xlabel('count')
plt.show()
```
